```python
import pandas as pd
import numpy as np
import datetime
```

```python
data= pd.read_excel('d.xlsx',sheet_name='Sheet1')
```

```python
data_1=data
```

```python
libur=pd.read_excel('d.xlsx',sheet_name='Sheet2')
```

```python
data_pp=data_1.dropna(subset=['OUTBND_CMPL_DTM', 'DLV_CMPL_DTM'])
```

```python
#membuat leadtime pp-rts
start = data_pp['OUTBND_CMPL_DTM'].values.astype('datetime64[D]')
end = data_pp['DLV_CMPL_DTM'].values.astype('datetime64[D]')
holiday = libur['holiday'].values.astype('datetime64[D]')
#Numpy function to find in between days
days = np.busday_count(start, end, weekmask='1111100', holidays=holiday)
```

```python
data_pp['PERF_PP_RTS']=days
```

```python
data_pp.head()
```

```python
data_ola=data_pp[['no','PERF_PP_RTS']]
```

```python
#menggabungkan nilai sla_shipping
data_2=pd.merge(data_1,data_ola, on='no', how='left')
```

```python
y = pd.DataFrame(data_2)
data_2.to_excel("data_out-dlv.xlsx")
```
